```python
#2.2 Den Median der Temperaturspalte ermitteln:
import pandas as pd

df = pd.read_excel('final_datensatz.xlsx')

median = df['temp'].median()

print('Der Median der Spalte {} ist {}'.format('temp', median))
```

```python
#2.3 Liniendiagramm der Temperatur erstellen. 
import pandas as pd
import matplotlib.pyplot as plt

# Lade die Daten aus der Excel-Datei in ein pandas DataFrame
data = pd.read_excel('final_datensatz.xlsx', parse_dates=['date'])

# Extrahiere Datum und Temperatur als separate Arrays
datum = data['date']
temperatur = data['temp']

# Erstelle das Liniendiagramm
plt.plot(datum, temperatur)

# Passe die Achsenbeschriftungen an
plt.xlabel('Datum')
plt.ylabel('Temperatur (°C)')

# Zeige das Diagramm an
plt.show()
```
